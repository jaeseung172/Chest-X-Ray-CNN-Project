### JSRT 데이터란 무엇인가

일본 방사선학회에서 제공하는 방사선 데이터입니다, 데이터는 1995년~1997년 사이에 수집/제작되었고, 흉부 X-Ray 결절 데이터입니다.

데이터는 PACS, Computer-Aided Diagnosis, ROC를 하는데 사용할 수 있습니다. 코니카 社의 LD-4500 과 LD-5500으로 현상 하였으며, 이미지는 2048 × 2048 사이즈의 이미지로 이루어져 있습니다. 

모든 사진은 엑스레이 데이터와 더불어서 CT스캔을 통해서 결절이라는 것을 판독하였고, 3명의 방사선사에 의해서 5등급으로 나누었습니다. 다음은 각 등급입니다.

등급|설명
---|---
1(엄청나게 교묘한) | 굉장히 교묘하게 되어 있어서, 사람이 발견하기 굉장히 힘듭니다
2(좀 교묘한) | 사람이 발견하기 힘듭니다
3(교묘한) | 그냥 봐서는 발견하기 힘듭니다
4(왠만하면 발견하기 쉬운) | 왠만하면 발견하기 쉽습니다
5(발견하기 쉬운) | 발견하기 쉽습니다

더 자세한 설명을 원한다면 다음 논문을 참고하여 주십시오
> Shiraishi J, Katsuragawa S, Ikezoe J, et al. "Development of a digital image database
for chest radiographs with and without a lung nodule: Receiver operating
characteristic analysis of radiologists' detection of pulmonary nodules," AMERICAN
JOURNAL OF ROENTGENOLOGY 174 (1): 71-74 JAN 2000

이 이미지는 사람이 육안으로 볼 수 없는 IMG 파일 확장자로 이루어져 있습니다, 그래서 이 이미지를 보실려면 IMAGE J라는 소프트웨어를 사용하여 볼 수 있습니다. 주소는 https://imagej.nih.gov/ij/ 이며 Download를 클릭하면 뷰어를 다운로드 받으실 수 있습니다.

해설 파일은 CLNDAT_E.txt 와 CNNDAT_E.txt 파일로 이루어져 있습니다, 위에서 말한 것 처럼 결절, 비 결절로 이루어져 있습니다. 샘플은 다음과 같습니다. 

**결절:**

JPCLN010.IMG 525 60 female 624 598 malignant r.upper lobe(S2) lung cancer

JPCLN011.IMG 520 68 male 664 386 benign r.upper lobe(S1) tuberculoma

이미지파일 결절사이즈(mm) 나이 성별 X좌표 Y좌표 등급 해부학적 위치 진단

**비 결절:**

### 데이터 레이블링

학습을 위하여 데이터를 레이블링 해야 합니다.

앞서 설명드린 CLNDAT_EN.txt와 CNNDAT_EN.txt 파일을 적절한 위치에 압축을 풀으셔서 밑의 스크립트를 실행하면 됩니다. 

학습할 데이터를 레이블링 해야 합니다, 여기서는 데이터의 크기가 그렇게 많진 않으니까 결절이 있다 없다 정도만 사용하는 것으로 하겠습니다. 

0이 True, 1이 False로 만들어 보도록 하겠습니다, 읽기 편하게 Key-Value 형식으로 작성하도록 하겠습니다.

그리고 컨블루전 네트워크만 사용할것이기 때문에 크게 좌표값을 줘서 학습하진 않겠습니다, V 사는 각각 표시를 해서 학습을 했지만, 아직은 그렇게 데이터가 썩어나는게 아니기 때문입니디다, 다만 학과 교수님들께서 데이터를 동원할 만한 능력만 있으시다면 솔루션을 사용하지 않고 학습하는건 누워서 떡먹기 입니다. 

```python
import json # Json 형태로 파일을 작성할 것이기 때문에 파일을 작성할 수 있도록 임포트합니다. 
CLNDAT = open("CLNDAT_EN.txt").readlines() #결절이 있습니다
CNNDAT = open("CNNDAT_EN.TXT").readlines() #결절이 없습니다
dat = {} #JSON 형식으로 만들어 보도록 하겠습니다
print("length of cln case: %d" % (len(CLNDAT)))
print("length of cnn case: %d" % (len(CNNDAT)))
print("sum of these case: %d" % (len(CLNDAT)+len(CNNDAT)))
```

Key:Value Data로 만들어 보겠습니다

```python
for i in range(0, len(CLNDAT)-1):
    dat[CLNDAT[i].split("\t")[0]] = 0
for d in range(0, len(CNNDAT)):
    dat[CNNDAT[d][0:12]] = 1
```

```python
dat # 데이터 확인
```

```python
len(dat) #데이터 갯수 확인
```

JSON 파일로 덤프하면.

```python
with open('lable.json', 'w') as dumper:
    json.dump(dat, dumper)
```
